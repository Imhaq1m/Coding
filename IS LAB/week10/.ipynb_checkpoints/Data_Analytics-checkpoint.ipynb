**Plotting with Python**

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting the line graph
plt.plot(x, y, marker='o', linestyle='-')  # 'o' marker for data points, '-' linestyle for a solid line
plt.title('Simple Line Graph')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)  # Show grid
plt.show()
```

In this example:


*   x contains the x-axis values.
*   y contains the corresponding y-axis values.


plt.plot(x, y) creates the line graph.
marker='o' specifies circular markers at data points.
linestyle='-' specifies a solid line connecting the data points.
plt.title(), plt.xlabel(), and plt.ylabel() are used to add a title and labels to the graph.
plt.grid(True) adds grid lines to the graph.

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 35, 30, 25, 40]

# Plotting the bar graph
plt.bar(categories, values, color='skyblue')
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Show grid lines on the y-axis
plt.show()
```

In this example:

* categories contains the labels for the x-axis (or categories).
* values contains the corresponding values for each category.
* plt.bar(categories, values) creates the bar graph.
* color='skyblue' sets the color of the bars to sky blue.
* plt.title(), plt.xlabel(), and plt.ylabel() are used to add a title and labels to the graph.
* plt.grid() adds grid lines to the graph. In this case, grid lines are added only on the y-axis with a dashed line style and reduced opacity.


**USING PENGUINS DATASET**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset from seaborn
penguins = sns.load_dataset('penguins')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(penguins.head())




```

```python
# Basic statistics
print("\nBasic statistics of the dataset:")
print(penguins.describe())
```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(penguins.isnull().sum())
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(penguins.head())
```

```python
# Data visualization
# Pairplot
sns.pairplot(penguins, hue='species')
plt.title('Pairplot of Penguins Dataset')
plt.show()


```

The sns.pairplot() function is a part of the Seaborn library, and it's used to create a grid of pairwise plots for a dataset. Each variable in the dataset is plotted against every other variable, resulting in a matrix of scatterplots by default.

In the case of sns.pairplot(penguins, hue='species'), the penguins DataFrame is passed as the first argument, indicating the dataset to be visualized. The hue='species' parameter is used to color the data points based on the values in the "species" column. This helps to distinguish between different species of penguins in the pair plot.

So, the resulting pair plot will display scatterplots of every pairwise combination of variables in the penguins dataset, with points colored according to the species of the penguins. This is useful for visually exploring relationships between multiple variables and understanding how they relate to each other within different species categories

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='bill_length_mm', data=penguins)
plt.title('Boxplot of Bill Length by Species')
plt.show()
```
