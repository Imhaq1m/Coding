```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the penguins dataset from seaborn
penguins = sns.load_dataset('penguins')
```

```python
penguins.dtypes
```

```python
#Exercise 1
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=penguins)

plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.title("Penguin Bill Length vs. Depth")
```

```python
#Exercise 2
sns.histplot(x='flipper_length_mm', hue='species', data=penguins, bins=30, multiple='stack')

plt.title('Flipper Lengths Among the Penguins')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Show grid lines on the y-axis
plt.show()
```

```python
#Exercise 3
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='body_mass_g', data=penguins)
plt.title('Boxplot of Body mass (g) by Species')
plt.show()
```

```python
#Exercise 4
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='flipper_length_mm', data=penguins)
plt.title('Violinplot of Flipper length (mm) by Species')
plt.show()
```

```python
#Exercise 5
selected_columns = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
sns.pairplot(data=penguins, vars=selected_columns, hue='species')
plt.title('Pairplot of Penguins Dataset')
plt.show()
```
