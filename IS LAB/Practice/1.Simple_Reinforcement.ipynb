```python
import gym

# Create the FrozenLake environment
env = gym.make('FrozenLake-v1', is_slippery=False)  # Non-slippery environment

# Set the number of episodes
num_episodes = 10

# Run the simulation
for i in range(num_episodes):
    state = env.reset()
    done = False
    total_reward = 0
    steps = 0

    # Print initial state
    print(f"Episode: {i+1}")
    env.render()
    print("-----")

    while not done and steps < 100:  # Add steps limit to avoid infinite loop
        if state in [0, 4, 8]:  # These states are aligned vertically on the left column
            action = 2  # Move right
        elif state in [1, 2, 6, 9, 10]:  # Middle states mostly require moving right
            action = 2  # Move right
        elif state == 3:  # From position 3 we need to move down to reach 7
            action = 1  # Move down
        elif state == 14:  # From position 14 move down to goal at 15
            action = 1  # Move down
        else:
            action = 0  # Default action is left (should not happen in this setup)

        new_state, reward, done, _ = env.step(action)

        # Normalize reward to 0 or 1
        normalized_reward = 1 if reward > 0 else 0

        total_reward += normalized_reward
        steps += 1

        # Print the results after action
        print(f"Step: {steps}")
        env.render()
        print("State: ", new_state)
        print("Reward: ", normalized_reward)
        print("Total Reward: ", total_reward)
        print("Terminated: ", done)
        print("-----")

        state = new_state  # Update state to new state

print("Simulation finished.\n")
```

```python
import gym
from gym import spaces

class CustomRewardEnv(gym.Wrapper):
    def __init__(self, env):
        super(CustomRewardEnv, self).__init__(env)

    def step(self, action):
        state, reward, done, info = self.env.step(action)
        # Define custom reward logic
        if not done and reward == 0:  # If the agent hasn't reached the goal and is on ice
            reward = -0.01  # Penalty for each step taken
        elif done and reward == 1:  # When the agent reaches the goal
            reward = 2  # Increase the reward for reaching the goal
        return state, reward, done, info

# Create the original FrozenLake environment
env = gym.make('FrozenLake-v1', is_slippery=False)

# Wrap the environment with the custom reward wrapper
env = CustomRewardEnv(env)

# Set the number of episodes
num_episodes = 10

# Run the simulation
for i in range(num_episodes):
    state = env.reset()
    done = False
    total_reward = 0
    steps = 0

    while not done:
        action = env.action_space.sample()  # Random action; replace with your policy if available
        new_state, reward, done, _ = env.step(action)
        total_reward += reward
        steps += 1

        if steps == 1 or done:
            print(f"Episode: {i+1}, Step: {steps}")
            print("State:", new_state, "Action:", action)
            print("Reward:", reward, "Total Reward:", total_reward)
            print("Terminated:", done)
            print("-----")

print("Simulation finished.\n")
```
