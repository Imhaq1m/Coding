In this example, the agent will follow a fixed policy (always move right unless it's not possible, then move down). The goal is to evaluate how effective this fixed policy is.

```python
import gym
import numpy as np

# Create the FrozenLake environment
env = gym.make('FrozenLake-v1', is_slippery=False)

def fixed_policy(state):
    return 2  # Always try to move right

# Evaluate the fixed policy
num_episodes = 100
total_rewards = []

for _ in range(num_episodes):
    state = env.reset()
    done = False
    total_reward = 0

    while not done:
        action = fixed_policy(state)  # Get action from the fixed policy
        new_state, reward, done, _ = env.step(action)
        total_reward += reward
        state = new_state

    total_rewards.append(total_reward)

print("Average Reward with Fixed Policy:", np.mean(total_rewards))
```

In this active learning example, we'll implement a simple Q-learning algorithm where the agent actively learns and updates its policy based on the environment's feedback.

```python
import gym
import numpy as np

# Create the FrozenLake environment
env = gym.make('FrozenLake-v1', is_slippery=False)

# Q-learning parameters
learning_rate = 0.8
discount_factor = 0.95
epsilon = 0.1  # Exploration rate
num_episodes = 1000
Q = np.zeros((env.observation_space.n, env.action_space.n))

# Q-learning algorithm
for _ in range(num_episodes):
    state = env.reset()
    done = False

    while not done:
        # Select action based on epsilon-greedy policy
        if np.random.random() < epsilon:
            action = env.action_space.sample()  # Explore
        else:
            action = np.argmax(Q[state, :])  # Exploit

        new_state, reward, done, _ = env.step(action)

        # Update Q-table
        old_value = Q[state, action]
        next_max = np.max(Q[new_state])

        new_value = (1 - learning_rate) * old_value + learning_rate * (reward + discount_factor * next_max)
        Q[state, action] = new_value

        state = new_state

# Test the learned policy
total_rewards = []
for _ in range(100):
    state = env.reset()
    done = False
    total_reward = 0

    while not done:
        action = np.argmax(Q[state, :])
        new_state, reward, done, _ = env.step(action)
        total_reward += reward
        state = new_state

    total_rewards.append(total_reward)

print("Average Reward with Learned Policy:", np.mean(total_rewards))
```
